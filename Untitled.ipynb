```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.randint(1, 50, 100)

# Histogram
sns.histplot(data, bins=10, kde=True, color='skyblue')
plt.title("Histogram of Data")
plt.show()
```

```python
sns.kdeplot(data, shade=True, color='#330000')
plt.title("KDE Plot of Data")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
tips = sns.load_dataset("tips")

# Create a histogram
sns.histplot(data=tips, x="tip", bins=20, kde=True, color="skyblue")

plt.title("Distribution of Tips")
plt.xlabel("Tip Amount ($)")
plt.ylabel("Frequency")
plt.show()
```

### Create a histogram for exam scores.

```python
names=[]
for i in range (1,21):
    name="name"+str(i)
    names.append(name)
    print(name)
indexs=[]
for i in range (1,21):
    index=i
    indexs.append(index)
print(indexs)

names1=pd.DataFrame(names,columns=['name'])   
scores=np.random.randint(50,100,20)
scores=pd.DataFrame(scores,columns=['score'])
scores
df=names1.join(scores)
df.index=indexs
df
```

```python
sns.histplot(data=df,x="score",kde=True,bins=8)
```

```python
sns.kdeplot(data=df, x="score")
```

```python
sns.kdeplot(data=df, x="score", fill=True, bw_adjust=0.8)
```
