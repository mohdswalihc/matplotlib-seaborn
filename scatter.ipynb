```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(1,55,10)
y=np.random.randint(1,55,10)
plt.scatter(x,y)
plt.show()
```

### subplot

```python
# plot 1
x=np.random.randint(1,55,10)
y=np.random.randint(1,55,10)
plt.scatter(x,y)
# plot 2
x1=np.random.randint(1,55,10)
y2=np.random.randint(1,55,10)
plt.scatter(x1,y2)
plt.show()
```

```python
# plot 1
x=np.random.randint(1,55,10)   
y=np.random.randint(1,55,10)
plt.scatter(x,y,c="black")
# plot 2
x1=np.random.randint(1,55,10)
y2=np.random.randint(1,55,10)
colors2=["#ff8000","#ffff00","#80ff00","#00ffff","#0080ff","#ff00ff","#bfff00","#ff00bf","#0000ff","#808080"]
plt.scatter(x1,y2,c=colors2)    # multiple coloring   "c"
plt.show()
```

```python
# plot 1
x=np.random.randint(1,55,10)   
y=np.random.randint(1,55,10)
size=np.random.randint(50,500,10)
plt.scatter(x,y,c="black",s=size)  ## sizing
# plot 2
x1=np.random.randint(1,55,10)
y2=np.random.randint(1,55,10)
colors2=["#ff8000","#ffff00","#80ff00","#00ffff","#0080ff","#ff00ff","#bfff00","#ff00bf","#0000ff","#808080"]
plt.scatter(x1,y2,c=colors2)    # multiple coloring   "c"
plt.show()
```

```python
# plot 1
x=np.random.randint(1,55,10)   
y=np.random.randint(1,55,10)
size=np.random.randint(50,500,10)
a=np.random.rand(10)
plt.scatter(x,y,c="black",s=size,alpha=a) # trancperacy
# plot 2
x1=np.random.randint(1,55,10)
y2=np.random.randint(1,55,10)
colors2=["#ff8000","#ffff00","#80ff00","#00ffff","#0080ff","#ff00ff","#bfff00","#ff00bf","#0000ff","#808080"]
plt.scatter(x1,y2,c=colors2)    
plt.show()
```

```python
##            color maping


# plot 1
x=np.random.randint(1,55,10)   
y=np.random.randint(1,55,10)
size=np.random.randint(50,500,10)
a=np.random.rand(10)
plt.scatter(x,y,c="black",s=size,alpha=a) # trancperacy
# plot 2
x1=np.random.randint(1,55,10)
y2=np.random.randint(1,55,10)
colors2=["#ff8000","#ffff00","#80ff00","#00ffff","#0080ff","#ff00ff","#bfff00","#ff00bf","#0000ff","#808080"]
plt.scatter(x1,y2,cmap='magma_r')  
plt.colorbar()
plt.show()
```
