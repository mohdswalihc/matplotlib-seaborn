```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

### pie 

```python
y=np.random.randint(1,50,5)
x=["A","B","C","D","E"]


plt.pie(y,labels=x)
plt.show()
```

```python
y=np.random.randint(1,50,5)
x=["A","B","C","D","E"]
explode=np.random.randn(5)

plt.pie(y,labels=x,explode=explode)   #### label
plt.show()
```

```python
y=np.random.randint(1,50,5)
x=["A","B","C","D","E"]
explode=[0,0,0,.1,.2]

plt.pie(y,labels=x,explode=explode)     ####    explode 
plt.show()
```

```python
y=np.random.randint(1,50,5)
x=["A","B","C","D","E"]
explode=[0,0,0,.1,.2]

plt.pie(y,labels=x,explode=explode,shadow=True)  #### shadow
plt.show()
```

```python
y=np.random.randint(1,50,5)
x=["A","B","C","D","E"]
explode=[0,0,0,.12,.12]
color=["red","green","blue","#00ffff","yellow"]
plt.pie(y,labels=x,explode=explode,colors=color) ### colors
plt.show()
```

```python
y=np.random.randint(1,50,5)
x=["A","B","C","D","E"]
explode=[0,0,0,.12,.12]
color=["red","green","blue","#00ffff","yellow"]
plt.pie(y,labels=x,explode=explode,colors=color,autopct="%1.1f%%") ### pecentage
plt.show()
```

```python
y=np.random.randint(1,50,5)
x=["A","B","C","D","E"]
explode=[0,0,0,.12,.12]
color=["red","green","blue","#00ffff","yellow"]
plt.pie(y,labels=x,explode=explode,colors=color,autopct="%1.1f%%")
plt.legend(loc="upper right", bbox_to_anchor=(1.2, 1))             #### legend
plt.show()
```
