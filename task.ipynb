```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Math', 'Science', 'English', 'History', 'Art']
boys_scores = [85, 90, 78, 88, 70]
girls_scores = [92, 85, 80, 75, 95]

x = np.arange(len(categories))  # X locations for groups
width = 0.35  # Width of the bars

# Create figure
plt.figure(figsize=(10,6))

# Bar plot (grouped bars)
plt.bar(x - width/2, boys_scores, label='Boys', color='skyblue', edgecolor='black', linewidth=2)
plt.bar(x + width/2, girls_scores, label='Girls', color='lightcoral', edgecolor='black', linewidth=2)

# Add titles and labels
plt.title('Average Subject Scores by Gender', fontsize=16, fontweight='bold', color='darkblue', loc='center', pad=15)
plt.xlabel('Subjects', fontsize=12, labelpad=10)
plt.ylabel('Scores', fontsize=12, labelpad=10)

# Customize x-axis ticks
plt.xticks(x, categories, rotation=30, fontsize=10)

# Add grid
plt.grid(True)

# Set y-axis limit
plt.ylim(0, 100)

# Add legend
plt.legend(title='Gender', loc='upper right', fontsize=10, title_fontsize=11)

# Add text labels on top of bars
for i in range(len(categories)):
    plt.text(x[i] - width/2, boys_scores[i] + 1, str(boys_scores[i]), ha='center', fontsize=9)
    plt.text(x[i] + width/2, girls_scores[i] + 1, str(girls_scores[i]), ha='center', fontsize=9)

# Add border and layout
plt.tight_layout()
plt.show()
```
