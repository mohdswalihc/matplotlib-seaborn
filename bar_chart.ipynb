```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

### bar chart

```python
label=["A","B","C","D"]
x=[20,30,40,60]
plt.bar(label,x)
```

```python
label=["A","B","C","D"]
x=[20,30,40,60]
plt.barh(label,x,height=.5)  ## height
```

```python
label=["A","B","C","D"]
x=[20,30,40,60]
plt.bar(label,x,width=.5)  #### width
```

```python
label=["A","B","C","D"]
x=[20,30,40,60]
plt.bar(label,x,width=.4,color="red",edgecolor="black")  #### color
```
