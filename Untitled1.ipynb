```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(100)
y = np.random.randn(100)

plt.figure(figsize=(2, 5))   # Correct spelling and syntax

plt.subplot(2, 2, 1)
plt.scatter(x, y, color='blue')

plt.subplot(2, 2, 2)
plt.scatter(x, y, color='green')

plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create grid points (X, Y)
x = np.arange(0, 5, 0.5)
y = np.arange(0, 5, 0.5)
X, Y = np.meshgrid(x, y)

# Define vector components (U, V)
U = np.cos(X)
V = np.sin(Y)

# Create quiver plot
plt.quiver(X, Y, U, V, color='blue')

# Add title and labels
plt.title('Quiver Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axis('equal')   # Equal scaling for both axes

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [25, 35, 20, 20]

# Pie chart
plt.pie(values,labels=categories,autopct='%1.1f%%')
plt.title('Category Distribution')
plt.show()
```

```python
explode = [0, 0.1, 0, 0]  # Highlight the second slice
colors = ['red', 'green', 'blue', 'orange']

plt.pie(values, labels=categories, autopct='%1.1f%%', explode=explode, colors=colors, startangle=9)
plt.title('Category Distribution')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Apply Seaborn style
plt.subplot(1,2,1)
sns.set_style("darkgrid")
plt.subplot(1,2,2)
# Example plot
sns.histplot([10, 20, 30, 40, 50])
sns.set_style("whitegrid")

# Example plot
sns.histplot([10, 20, 30, 40, 50])
plt.title("Using 'darkgrid' Style")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 30, 40, 50]

for style in ['darkgrid', 'whitegrid', 'dark', 'white']:
    sns.set_style(style)
    plt.figure()
    sns.histplot(data)
    plt.title(f"Style: {style}")
    plt.show()
```
