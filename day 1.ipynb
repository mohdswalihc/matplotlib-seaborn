### Draw a histogram .

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Create some numeric data
data = np.random.randint(1, 100, 50)  # 50 random integers between 1 and 99

print("Data:", data)

# Create histogram
sns.histplot(data, bins=100)  # bins=10 divides data into 10 ranges
plt.title("Histogram of Data")
plt.xlabel("Value")
plt.grid(True)
plt.ylabel("Frequency")
plt.show()
```

```python
np.bincount((data))
```

#### Plot a line chart for numbers 1–10.


```python
import matplotlib.pyplot as plt

# Data
x = list(range(1, 11))  # Numbers 1 to 10
 # For simplicity, y = x

# Plot
plt.plot(x, marker='o', linestyle='-', color='blue')  # Line with markers
plt.title("Line Chart of Numbers 1 to 10")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)  # Optional grid
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = list(range(1, 11))  # Numbers 1 to 10
y = x  # For simplicity, y = x

# Plot
plt.plot(x, y, marker='o', linestyle='-', color='blue')  # Line with markers
plt.title("Line Chart of Numbers 1 to 10")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)  # Optional grid
plt.show()
```

### Plot a sine and cosine curve on the same chart with labels.

```python
s = np.linspace(0, 2*np.pi, 100)
s
```

```python
x=np.sin(s)
y=np.cos(s)

plt.plot(s,x,label="sine")
plt.plot(s,y,label="cos")
plt.legend()
plt.show()
```

### Customize the chart by adding title, xlabel, ylabel.

```python
x=np.sin(s)
y=np.cos(s)

plt.plot(s,x,label="sine")
plt.plot(s,y,label="cos")
plt.legend()
plt.title('Sine and Cosine Waves')
plt.xlabel("Angle")
plt.ylabel('Value')
plt.show()
```
