```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

### Heatmaps & Clustermaps


```python
data = pd.DataFrame({
    "Math": [90, 50, 75, 40],
    "Science": [85, 60, 70, 35],
    "English": [80, 65, 90, 50],
    "History": [70, 55, 95, 0]
}, index=["A", "B", "C", "D"])  # student names as index

print(data)
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(data=data)
yticklabels=''
xticklabels='auto'
plt.show()
```

```python
sns.heatmap(data.corr(),annot=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = data.corr()
print(corr)

sns.heatmap(corr, annot=True, cmap="RdYlBu", center=0)
plt.title("Subject Correlation Heatmap")
plt.show()
```

### Create a cluster map to group similar rows/columns.


```python
sns.clustermap(
    data,                 # The dataset
    cmap="coolwarm",      # Color palette
    annot=True,           # Show data values
    figsize=(8, 5),       # Adjust figure size
    standard_scale=1      # Normalize by columns for fair comparison
)
plt.title("Student Performance Cluster Map", pad=50)
plt.show()
```

```python
sns.clustermap(
    data,
    cmap="YlGnBu",
    annot=True,
    linewidths=0.5,
    figsize=(5, 5),
    method="ward",        # Clustering method (ward, average, single)
    metric="euclidean",   # Distance metric
)
plt.show()
```

```python
# Create a 55 random matrix with values between 0 and 1
data = np.random.rand(5, 5)

# Convert to DataFrame for better labeling
df = pd.DataFrame(data, 
                  columns=['A', 'B', 'C', 'D', 'E'], 
                  index=['R1', 'R2', 'R3', 'R4', 'R5'])

print(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Calculate correlation matrix
corr = df.corr(numeric_only=True)  # ensures only numeric columns are included

# Step 2: Create correlation heatmap
sns.heatmap(
    corr,
    annot=True,        # show correlation values
    cmap="coolwarm",   # color range: blue (negative)  red (positive)
    linewidths=0.5,    # grid lines
    cbar=True          # show color bar (acts as legend)
)

# Step 3: Add title
plt.title("Correlation Heatmap of Numeric Features", fontsize=12, pad=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
students = ['A', 'B', 'C', 'D', 'E']
marks = [70, 85, 60, 90, 75]

# Find mean
mean_marks = np.mean(marks)
print("Mean:", mean_marks)
```

```python
plt.bar(students, marks, color='skyblue')
plt.axhline(mean_marks, color='red')
plt.title("Marks of Students with Mean Line")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.legend()
plt.show()
```
