### Visualize spread with violin and strip plots.
1 Violin Plot

A violin plot shows the distribution of a numeric variable across different categories.
Its like a boxplot + density plot: you can see the spread, median, and shape of the data distribution.
The width of the violin represents how dense the data is at different values.
Use case: To visualize both the distribution and summary statistics of a numeric variable across categories.

2 Strip Plot

A strip plot shows individual data points along a categorical axis.
Each point represents one observation; points can be jittered to avoid overlapping.
Often used to see all raw data points rather than just summary statistics.
Use case: To visualize every observations spread across categories, especially when there arent too many points.

```python
data = pd.DataFrame({
    "Class": ["A", "A", "A", "B", "B", "B", "C", "C", "C"],
    "Score": [85, 78, 90, 88, 92, 81, 79, 95, 87]
})
data
```

```python
sns.violinplot(data=data, x='Class', y='Score', palette='Pastel1')
```

```python
sns.stripplot(data=data, x='Class', y='Score')
```

```python
plt.scatter(data=data,x='Class',y='Score')
```

### Plot a bar chart of average marks by subject.
Bar Chart

A bar chart is used to compare values across categories.
Each bars height (or length) represents a value associated with a category.
Can be used for counts (like countplot) or aggregated numeric values (like average, sum, etc.).

Average Marks by Subject

If you have marks for multiple students in different subjects, you can aggregate the marks (like compute the mean) for each subject.
Then, a bar chart will show which subject has the highest or lowest average score.

```python
data = pd.DataFrame({
    "Student": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Math": [85, 90, 78, 88, 92],
    "Science": [80, 85, 88, 90, 86],
    "English": [75, 82, 79, 85, 88]
})
data
```

```python
avg_mark=data[['English',"Science","Math"]].mean()
avg_mark
```

```python
avg_mark.plot(kind="bar")
```

```python
avg_mark.plot(kind="pie", color=["skyblue","lightgreen","pink"])
avg_mark
```

### Overlay strip plot on top of a boxplot for salary data.

```python
data = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Finance", "Finance", "IT", "HR", "Finance", "IT"],
    "Salary": [50000, 55000, 70000, 72000, 60000, 58000, 75000, 52000, 61000, 73000]
})
```

```python
sns.boxplot(data=data, x="Department", y="Salary", palette="Pastel1")
plt.title("Salary Distribution by Department (Boxplot)")
plt.show()
```

```python
sns.stripplot(data=data, x="Department", y="Salary", color="black", size=8, jitter=True)
plt.title("Salary Spread by Department (Strip Plot)")
plt.show()
```

```python
sns.boxplot(data=data, x="Department", y="Salary", palette="Pastel1")
sns.stripplot(data=data, x="Department", y="Salary", color="black", size=6, jitter=True)
plt.title("Salary Distribution with Individual Points")
plt.show()
```
